```python
import pandas as pd
import time
import math
import numpy as np
import string as str
import openpyxl
from datetime import datetime
import sys
import statsmodels.api as sm
import matplotlib.pyplot as plt

%matplotlib inline

infile =  r"C:\Users\HSASS\OneDrive - Wipro\Desktop\AIR\AIR_BFSI_today.csv"
table = pd.read_csv(infile, engine='python',sep=',', quotechar='"', error_bad_lines=False)

table.head()
```

```python
table.describe()
```

```python
df = table[['INDENT_CREATED_ON','TOTAL_REQUIRED']]
```

```python
df.head()
```

```python
df.columns = ['Month','Total']
```

```python
df.head()
```

```python
df['Month'] = pd.to_datetime(df['Month'])
df.head()
```

```python
df.set_index('Month',inplace=True)
df.head()
```

```python
ax = df['Total'].plot(kind='bar')
#df2.groupby(['embark_town']).count().pclass
ax.set_ylabel('count')
ax.set_xlabel('Months')
plt.xticks(rotation=0)
```

```python
timeseries = df ['Total']
timeseries.rolling(12).mean().plot(label='12 Month Rolling Mean')
timeseries.rolling(12).std().plot(label='12 Month Rolling Std')
timeseries.plot()
plt.legend()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df['Total'], freq=12)  
figure = plt.figure()  
figure = decomposition.plot()  
figure.set_size_inches(15, 8)
```

```python
from statsmodels.tsa.stattools import adfuller
test_result = adfuller(df['Total'])
print ('ADF Test:')
labels = ['ADF Statistic','p-value','No. of Lags Used','Number of Observations Used']

for value,label in zip(test_result,labels):
    print (label+': '+str(value))
if test_result [1] <= 0.05:
    print ("Reject null hypothesis and data is stationary")
else:
    print ("Fail to reject H0 thereby data is non-stationary ")
```
